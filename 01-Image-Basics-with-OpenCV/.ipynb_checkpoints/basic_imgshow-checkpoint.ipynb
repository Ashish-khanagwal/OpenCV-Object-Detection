```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
img = cv2.imread('../Computer-Vision-with-Python/DATA/00-puppy.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
# MATPLOTLIB --> RGB (RED GREEN BLUE)
# OPENCV --> BGR (BLUE GREEN RED)
# So, now we have to fix this thing for opnecv and fortunately OpenCv has functionality to do so.
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('../Computer-Vision-with-Python/DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

```python
# RESIZING AN IMAGE
fix_img.shape
```

```python
new_img = cv2.resize(fix_img, (500,500))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.5 # 50% of width
h_ratio = 0.5 # 50% of height
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
new_img = cv2.flip(fix_img,-1)
plt.imshow(new_img)
```
