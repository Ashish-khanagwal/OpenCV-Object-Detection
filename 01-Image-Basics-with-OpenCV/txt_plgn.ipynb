```python
# In this we'll learn about drawing Text and Polygons on Image
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros((512, 512, 3), dtype = np.int16)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text = 'Hello', org = (200, 200), fontFace = font, fontScale = 4, color = (255, 255, 255), thickness = 3, lineType = cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
# Drawing a polygon on image
black_img = np.zeros((512,512,3), dtype = np.int32)
plt.imshow(black_img)
```

```python
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype = np.int32)
```

```python
vertices.shape
```

```python
vertices
```

```python
pts = vertices.reshape(-1, 1, 2)
pts
```

```python
pts.shape
```

```python
cv2.polylines(black_img, [pts], isClosed = True, color = (255, 255, 0), thickness = 5)
plt.imshow(black_img)
```
