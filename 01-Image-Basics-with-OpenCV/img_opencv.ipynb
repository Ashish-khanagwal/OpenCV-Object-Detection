```python
# In this we'll gonna learn about creating/drawing on images

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros((512,512,3), dtype = np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
# Drawing a rectangle on above blank image
cv2.rectangle(blank_img, pt1 = (250,100), pt2 = (400,150), color = (0, 255,0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
# Drawing an square on image
cv2.rectangle(blank_img, pt1 = (200,200), pt2 = (300,300), color = (0, 0, 255), thickness = 10)
```

```python
plt.imshow(blank_img)
```

```python
# Drawing an circle on image
cv2.circle(blank_img, center = (100,100), radius = 50, color = (255, 0, 0), thickness = 10)
```

```python
plt.imshow(blank_img)
```

```python
# Upto this point we are drawing shapes on images but hollow, well if we want to draw filled images then set "thickness = -1"
cv2.circle(blank_img, center = (400, 400), radius = 50, color = (255, 0, 0), thickness = -1)
```

```python
plt.imshow(blank_img)
```

```python
# Drawing an line on image
cv2.line(blank_img, pt1 = (0, 0), pt2 = (512, 512), color = (110, 255, 255), thickness = 5)
plt.imshow(blank_img)
```
