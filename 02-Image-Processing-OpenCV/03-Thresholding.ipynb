```python
# In this, we'll learn about Thresholding. Basically Thresholding will convert an image to consist of only two values White or Black.
# In thresholding, each pixel value is compared with the threshold value. If the pixel value is smaller than the threshold, it is set to 0, otherwise, it is set to a maximum value (generally 255)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../Computer-Vision-with-Python/DATA/rainbow.jpg', 0)
# To read this image as grayscale, we just simply put '0' at the end after reading the image
plt.imshow(img, cmap = 'gray')
```

```python
img.max()
```

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
```

```python
ret
```

```python
plt.imshow(thresh1, cmap = 'gray')
```

```python
img2 = cv2.imread('../Computer-Vision-with-Python/DATA/crossword.jpg', 0)
plt.imshow(img2, cmap = 'gray')
```

```python
def show_img(img2):
    fig = plt.figure(figsize = (15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img2, cmap = 'gray')
```

```python
show_img(img2)
```

```python
ret, thresh = cv2.threshold(img2, 150, 255, cv2.THRESH_BINARY)
```

```python
show_img(thresh)
```

```python
# Now, we will look into Adaptive Thresholding.
# Adaptive thresholding is the method where the threshold value is calculated for smaller regions. This leads to different threshold values for different regions with respect to the change in lighting.
```

```python
th = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8)

show_img(th)
```

```python
blended_img = cv2.addWeighted(thresh, 0.6, th, 0.4, 0)
show_img(blended_img)
```
