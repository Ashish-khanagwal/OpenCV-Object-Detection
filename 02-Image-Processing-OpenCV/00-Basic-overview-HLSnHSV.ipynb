```python
# In this we will learn about converting RGB/BGR to HLS or HSV
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../Computer-Vision-with-Python/DATA/dog_backpack.jpg')
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
# As we've converted BGR to RGB, in the same way we can convert from RGB to HSV or HLS
cvt_img = cv2.cvtColor(fix_img, cv2.COLOR_RGB2HLS) # Converted from RGB to HLS
plt.imshow(cvt_img)
```

```python
cvt_img = cv2.cvtColor(fix_img, cv2.COLOR_RGB2HSV) # Converted from RGB to HSV
plt.imshow(cvt_img)
```
